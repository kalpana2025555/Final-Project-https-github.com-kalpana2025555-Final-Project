```python
import zipfile
import os
import cv2

# Path to ZIP file
zip_path = r"C:\Users\kalpana\Downloads\images-20250217T092357Z-001.zip"
extract_folder = r"C:\Users\kalpana\Downloads\extracted_images\images"

# Extract ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# Get a list of image files in the extracted folder
image_files = [f for f in os.listdir(extract_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Ensure there are images
if not image_files:
    raise FileNotFoundError("No images found in the extracted ZIP folder.")

# Read the first image
image_path = os.path.join(extract_folder, image_files[0])
image = cv2.imread(image_path)

# Ensure image is loaded
if image is None:
    raise FileNotFoundError(f"Error loading image: {image_path}")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Proceed with face detection or processing...
print(f"Successfully loaded and processed {image_path}")
```

```python
import dlib
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern

# Load Dlib's pre-trained face detector and landmark predictor
face_detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("C:\\Users\\kalpana\\Downloads\\shape_predictor_68_face_landmarks (1).dat")

# Detect faces
detections = face_detector(gray)

for face in detections:
    x, y, w, h = face.left(), face.top(), face.width(), face.height()
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Extract facial landmarks
    landmarks = predictor(gray, face)
    for n in range(68):
        x_lm, y_lm = landmarks.part(n).x, landmarks.part(n).y
        cv2.circle(image, (x_lm, y_lm), 2, (255, 0, 0), -1)

# Apply Histogram Equalization
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
equalized = clahe.apply(gray)

# Normalize pixel values
normalized = gray.astype(np.float32) / 255.0

# Extract LBP Features
radius = 1
n_points = 8 * radius
lbp = local_binary_pattern(gray, n_points, radius, method='uniform')

# Display Results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title("Face Detection & Landmarks")
axs[1].imshow(equalized, cmap='gray')
axs[1].set_title("Histogram Equalization")
axs[2].imshow(lbp, cmap='gray')
axs[2].set_title("Local Binary Patterns")
plt.show()
```
